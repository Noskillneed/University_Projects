**Outlier Detection and Relation**

Visual inspection: To answer my questions, I used a scatterplot to find outliers and relations by comparing 'Quality of Sleep' with 'Sleep Duration' and 'Stress_lvl' as well as comparing 'Sleep Duration' with 'Stress_lvl':








Here are some insights I reached:

● There are many outliers in this dataset since it is made up.

● There is a negative correlation between 'Quality of Sleep' and 'Stress_lvl' (As stress gets increase sleep quality decrease). Similarly, there is a negative correlation between 'Sleep Duration' and 'Stress_lvl' (As stress gets increase sleep duration decrease).

● There is a positive correlation between 'Quality of Sleep' and 'Sleep Duration' (As sleep duration increase sleep quality  increase)

● A viscous cycle happens here. As a person gets more stressed, their sleep duration shortens thus reducing the sleep quality of the individual which in turns could increase stress levels over a long period of time.
